# 19. Limits of Sequences

---

## **1. Calculate:**

### **(a)**
$$
\lim_{n \to \infty} \frac{n^2 + 3n}{2n^2 - 2n}
$$

### **Solution:**
Divide the numerator and denominator by \( n^2 \):

$$
\lim_{n \to \infty} \frac{\frac{n^2}{n^2} + \frac{3n}{n^2}}{\frac{2n^2}{n^2} - \frac{2n}{n^2}}
= \lim_{n \to \infty} \frac{1 + \frac{3}{n}}{2 - \frac{2}{n}}
$$

Since \( \frac{3}{n} \to 0 \) and \( \frac{2}{n} \to 0 \) as \( n \to \infty \), we get:

$$
\frac{1 + 0}{2 - 0} = \frac{1}{2}
$$

\[
\boxed{\frac{1}{2}}
\]

---

### **(b)**
$$
\lim_{n \to \infty} \frac{(2n + 3)^3}{n^3 - 1}
$$

### **Solution:**
Expanding \( (2n + 3)^3 \):

$$
(2n + 3)^3 = 8n^3 + 12n^2(3) + 6n(9) + 27 = 8n^3 + 36n^2 + 54n + 27
$$

Dividing numerator and denominator by \( n^3 \):

$$
\lim_{n \to \infty} \frac{8 + \frac{36}{n} + \frac{54}{n^2} + \frac{27}{n^3}}{1 - \frac{1}{n^3}}
$$

Since \( \frac{36}{n}, \frac{54}{n^2}, \frac{27}{n^3} \to 0 \) and \( \frac{1}{n^3} \to 0 \), we get:

$$
\frac{8 + 0}{1 - 0} = 8
$$

\[
\boxed{8}
\]

## **2. Prove using the squeeze theorem:**
$$
\lim_{n \to \infty} \frac{\sin(n)}{n}
$$

### **Solution:**
Since \( -1 \leq \sin(n) \leq 1 \), dividing all sides by \( n > 0 \):

$$
-\frac{1}{n} \leq \frac{\sin(n)}{n} \leq \frac{1}{n}
$$

Taking the limit as \( n \to \infty \):

$$
\lim_{n \to \infty} -\frac{1}{n} = 0, \quad \lim_{n \to \infty} \frac{1}{n} = 0
$$

By the **squeeze theorem**, we conclude:

$$
\lim_{n \to \infty} \frac{\sin(n)}{n} = 0
$$

\[
\boxed{0}
\]
