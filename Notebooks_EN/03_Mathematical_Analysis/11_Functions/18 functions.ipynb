# 18. Functions

## 1. Draw in a single Geogebra notebook the following functions:

- \( f(x) = x^2 \)
- \( g(x) = \sqrt{x} \)
- \( h(x) = \frac{1}{x} \)
- \( j(x) = \sin(x) \)

Find the values of all the above functions at \( x = 2 \).

### Solution:

\[
f(2) = 2^2 = 4
\]

\[
g(2) = \sqrt{2} \approx 1.414
\]

\[
h(2) = \frac{1}{2} = 0.5
\]

\[
j(2) = \sin(2) \approx 0.909
\]

\[
\boxed{(f(2) = 4, g(2) \approx 1.414, h(2) = 0.5, j(2) \approx 0.909)}
\]

## 2. Let \( f(x) = 3x - 1 \) and \( g(x) = \sqrt{x} \). Find:

- \( f(g(x)) \)
- \( g(f(x)) \)
- \( f(f(x)) \)
- \( g(g(x)) \)

### Solution:

#### 1. \( f(g(x)) \)

\[
f(g(x)) = f(\sqrt{x}) = 3\sqrt{x} - 1
\]

#### 2. \( g(f(x)) \)

\[
g(f(x)) = g(3x - 1) = \sqrt{3x - 1}
\]

#### 3. \( f(f(x)) \)

\[
f(f(x)) = f(3x - 1) = 3(3x - 1) - 1 = 9x - 3 - 1 = 9x - 4
\]

#### 4. \( g(g(x)) \)

\[
g(g(x)) = g(\sqrt{x}) = \sqrt{\sqrt{x}} = x^{1/4}
\]

\[
\boxed{(f(g(x)) = 3\sqrt{x} - 1, g(f(x)) = \sqrt{3x - 1}, f(f(x)) = 9x - 4, g(g(x)) = x^{1/4})}
\]

## 3. Let \( f(x) = e^x \) and \( g(x) = \ln(x) \). Check: \( f(g(x)) \) and \( g(f(x)) \). What do you notice?

### Solution:

\[
f(g(x)) = e^{\ln(x)} = x
\]

\[
g(f(x)) = \ln(e^x) = x
\]

Since \( f(g(x)) = x \) and \( g(f(x)) = x \), they are **inverse functions** of each other.

\[
\boxed{(f(g(x)) = x, g(f(x)) = x)}
\]

## 4. We have function \( f = (1,7), (2,9), (3,11) \). Give inverse function \( f^{-1} \).

### Solution:

Swapping \( x \) and \( y \) values:

\[
f^{-1} = (7,1), (9,2), (11,3)
\]

\[
\boxed{f^{-1} = (7,1), (9,2), (11,3)}
\]

## 5. We have function \( f = (1,7), (2,7), (3,11) \). Give inverse function \( f^{-1} \).

### Solution:

Since \( f(1) = 7 \) and \( f(2) = 7 \), the function is **not one-to-one**, meaning \( f^{-1} \) is not a function.

\[
\boxed{\text{Inverse does not exist (not one-to-one)}}
\]

---

## 6. We have function \( f(x) = x - 1 \). Give inverse function \( f^{-1} \). Show both functions on the same Geogebra notebook.

### Solution:

To find \( f^{-1} \), swap \( x \) and \( y \):

\[
y = x - 1
\]

Solving for \( x \):

\[
x = y + 1
\]

Thus,

\[
f^{-1}(x) = x + 1
\]

\[
\boxed{f^{-1}(x) = x + 1}
\]
