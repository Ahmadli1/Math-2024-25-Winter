# 20. Limits of Real Functions

---

## **1. Compute:**

### **(a)**
$$
\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3}
$$

### **Solution:**
Divide the numerator and denominator by \( x^4 \):

$$
\lim_{x \to \infty} \frac{\frac{x^3}{x^4} + \frac{2x^2}{x^4}}{\frac{x^4}{x^4} - \frac{3x^3}{x^4}}
= \lim_{x \to \infty} \frac{\frac{1}{x} + \frac{2}{x^2}}{1 - \frac{3}{x}}
$$

Since \( \frac{1}{x} \to 0 \) and \( \frac{2}{x^2} \to 0 \) as \( x \to \infty \), we get:

$$
\frac{0 + 0}{1 - 0} = 0
$$

\[
\boxed{0}
\]

## **2. Find:**
$$
\lim_{x \to 0} \frac{\sin(3x)}{2x + 1}
$$

### **Solution:**
Using the **small-angle approximation** \( \sin(3x) \approx 3x \) for small \( x \):

$$
\lim_{x \to 0} \frac{\sin(3x)}{2x + 1} = \lim_{x \to 0} \frac{3x}{2x + 1}
$$

Substituting \( x = 0 \):

$$
\frac{3(0)}{2(0) + 1} = \frac{0}{1} = 0
$$

\[
\boxed{0}
\]

## **3. Find the asymptotes of the function:**

### **(a)**
$$
f(x) = \frac{x^2 - 1}{x^2 + 1}
$$

#### **Horizontal Asymptote:**
For large \( x \), divide numerator and denominator by \( x^2 \):

$$
\lim_{x \to \infty} \frac{x^2 - 1}{x^2 + 1} = \frac{1 - \frac{1}{x^2}}{1 + \frac{1}{x^2}}
$$

Since \( \frac{1}{x^2} \to 0 \), we get:

$$
\frac{1 - 0}{1 + 0} = 1
$$

Thus, the **horizontal asymptote** is:

\[
\boxed{y = 1}
\]

#### **Vertical Asymptote:**
Since the denominator \( x^2 + 1 \) is never **zero**, there are **no vertical asymptotes**.

---

### **(b)**
$$
g(x) = \frac{\sin(x)}{x^2 + 1}
$$

#### **Horizontal Asymptote:**
For large \( x \), since \( |\sin(x)| \leq 1 \), we get:

$$
\left| \frac{\sin(x)}{x^2 + 1} \right| \leq \frac{1}{x^2 + 1}
$$

Since \( \frac{1}{x^2 + 1} \to 0 \) as \( x \to \infty \), we conclude:

$$
\lim_{x \to \infty} g(x) = 0
$$

Thus, the **horizontal asymptote** is:

\[
\boxed{y = 0}
\]

#### **Vertical Asymptote:**
Since the denominator \( x^2 + 1 \) is never **zero**, there are **no vertical asymptotes**.
