## **1. Compute derivatives of functions:**

### **(a)** \( y(x) = -3x + 3 \)
$$
y'(x) = -3
$$

### **(b)** \( y(x) = \pi x + \sin(1) \)
$$
y'(x) = \pi
$$

### **(c)** \( y(x) = 4 + \sin(2) \)
$$
y'(x) = 0
$$

### **(d)** \( y(x) = 2x^3 - 3x^2 + 8x - 9 \)
$$
y'(x) = 6x^2 - 6x + 8
$$

### **(e)** \( y(x) = 6x^{1/3} \)
Using the power rule:

$$
y'(x) = 6 \times \frac{1}{3} x^{-2/3} = 2x^{-2/3} = \frac{2}{x^{2/3}}
$$

### **(f)** \( y(x) = \sqrt{x} \)
$$
y'(x) = \frac{1}{2\sqrt{x}}
$$

### **(g)** \( y(x) = \cos(x) + \sin(x) \)
$$
y'(x) = -\sin(x) + \cos(x)
$$

### **(h)** \( y(x) = 2\sin(x)\cos(x) \)
Using the identity \( 2\sin(x)\cos(x) = \sin(2x) \):

$$
y'(x) = 2\cos(2x)
$$

### **(i)** \( y(x) = x\sin(x) \)
Using the product rule:

$$
y'(x) = x\cos(x) + \sin(x)
$$

### **(j)** \( y(x) = (x+1)(x+1) \)
Rewriting:

$$
y(x) = (x+1)^2
$$

Using the chain rule:

$$
y'(x) = 2(x+1)
$$

### **(k)** \( y(x) = \frac{x}{x+1} \)
Using the quotient rule:

$$
y'(x) = \frac{(x+1) \cdot 1 - x \cdot 1}{(x+1)^2} = \frac{(x+1 - x)}{(x+1)^2} = \frac{1}{(x+1)^2}
$$

### **(l)** \( y(x) = (x+1) \exp(x) \)
Using the product rule:

$$
y'(x) = (x+1) \exp(x) + \exp(x) = (x+2) \exp(x)
$$

### **(m)** \( y(x) = \sin(x^2) \)
Using the chain rule:

$$
y'(x) = \cos(x^2) \cdot 2x = 2x\cos(x^2)
$$

### **(n)** \( y(x) = \exp(-2x) \)
Using the chain rule:

$$
y'(x) = \exp(-2x) \cdot (-2) = -2\exp(-2x)
$$

### **(o)** \( y(x) = \frac{1}{\sin(x+1)} \)
Using the chain and reciprocal rule:

$$
y'(x) = -\frac{\cos(x+1)}{\sin^2(x+1)}
$$

### **(p)** \( y(x) = \sqrt{2x+1} \)
Using the chain rule:

$$
y'(x) = \frac{1}{2\sqrt{2x+1}} \cdot 2 = \frac{1}{\sqrt{2x+1}}
$$

## **2. Prove:**
$$
\frac{d}{dx} (\ln(\sin(x))) = \cot(x)
$$

### **Solution:**
Using the chain rule:

$$
\frac{d}{dx} \ln(\sin(x)) = \frac{1}{\sin(x)} \cdot \cos(x)
$$

Since \( \frac{\cos(x)}{\sin(x)} = \cot(x) \), we obtain:

$$
\boxed{\frac{d}{dx} (\ln(\sin(x))) = \cot(x)}
$$

## **3. For** \( f(x) = \cos(x) \) **, verify that** \( f''(x) = -f(x) \).

### **Solution:**
First derivative:

$$
f'(x) = -\sin(x)
$$

Second derivative:

$$
f''(x) = -\cos(x) = -f(x)
$$

\[
\boxed{f''(x) = -f(x)}
\]

## **4. Using l’Hôpital’s Rule, find the improper limits:**

### **(a)**
$$
\lim_{x \to 0} \frac{\sin x}{x}
$$

Using L'Hôpital’s Rule:

$$
\lim_{x \to 0} \frac{\cos x}{1} = 1
$$

\[
\boxed{1}
\]

---

### **(b)**
$$
\lim_{x \to \infty} \frac{\ln x}{x}
$$

Using L'Hôpital’s Rule:

$$
\lim_{x \to \infty} \frac{\frac{1}{x}}{1} = \lim_{x \to \infty} \frac{1}{x} = 0
$$

\[
\boxed{0}
\]

---

### **(c)**
$$
\lim_{x \to -\infty} \exp(x)
$$

Since \( e^x \to 0 \) as \( x \to -\infty \), we get:

\[
\boxed{0}

## **5. In physics, the position of a particle is given by** \( x(t) = 3t^2 - 6t + 1 \).  
Find the velocity \( V(t) = x'(t) \) and acceleration \( a(t) = V'(t) \) of the particle at \( t = 2 \).

### **Solution:**
First derivative (velocity):

$$
V(t) = x'(t) = 6t - 6
$$

At \( t = 2 \):

$$
V(2) = 6(2) - 6 = 12 - 6 = 6
$$

Second derivative (acceleration):

$$
a(t) = V'(t) = 6
$$

At \( t = 2 \):

$$
a(2) = 6
$$

\[
\boxed{V(2) = 6, \quad a(2) = 6}
\]
