## **1. Solve the following first-order ordinary differential equations:**

### **(a)**  
$$ y'(x) = y(x) $$

### **Solution:**
Rearrange to separate variables:

$$
\frac{dy}{dx} = y
$$

$$
\frac{dy}{y} = dx
$$

Integrating both sides:

$$
\int \frac{dy}{y} = \int dx
$$

$$
\ln |y| = x + C
$$

Exponentiating both sides:

$$
y = Ce^x
$$

\[
\boxed{y = Ce^x}
\]

---

### **(b)**  
$$ y'(x) = \frac{1}{2y(x)} $$

### **Solution:**
Rewriting:

$$
2y \frac{dy}{dx} = 1
$$

Separating variables:

$$
2y \, dy = dx
$$

Integrating both sides:

$$
\int 2y \, dy = \int dx
$$

$$
y^2 = \frac{x}{2} + C
$$

\[
\boxed{y = \pm\sqrt{\frac{x}{2} + C}}
\]

---

## **2. Solve the first-order ordinary differential equations using the method of separation of variables for** \( x \) **and** \( y = y(x) \):

### **(a)**  
$$ \frac{dy}{dx} = \frac{x}{y} $$

### **Solution:**
Separating variables:

$$
y \, dy = x \, dx
$$

Integrating:

$$
\int y \, dy = \int x \, dx
$$

$$
\frac{y^2}{2} = \frac{x^2}{2} + C
$$

\[
\boxed{y^2 = x^2 + 2C}
\]

---

### **(b)**  
$$ \frac{dy}{dx} = \frac{y}{x} $$

### **Solution:**
Separating variables:

$$
\frac{dy}{y} = \frac{dx}{x}
$$

Integrating:

$$
\int \frac{dy}{y} = \int \frac{dx}{x}
$$

$$
\ln |y| = \ln |x| + C
$$

Exponentiating both sides:

$$
y = Cx
$$

\[
\boxed{y = Cx}
\]

---

### **(c)**  
$$ \frac{dy}{dx} = xy $$

### **Solution:**
Separating variables:

$$
\frac{dy}{y} = x \,dx
$$

Integrating:

$$
\int \frac{dy}{y} = \int x \,dx
$$

$$
\ln |y| = \frac{x^2}{2} + C
$$

Exponentiating both sides:

$$
y = e^C e^{x^2/2} = C' e^{x^2/2}
$$

\[
\boxed{y = C e^{x^2/2}}
\]

## **3. Solve the second-order ordinary differential equations:**

### **(a)**  
$$ y''(x) + y'(x) = 0, \quad \text{with boundary conditions } y(0) = 2, \, y'(0) = -1 $$

### **Solution:**
Solve the characteristic equation:

$$
r^2 + r = 0
$$

$$
r(r + 1) = 0
$$

Roots:

$$
r = 0, \quad r = -1
$$

Thus, the general solution:

$$
y(x) = C_1 + C_2 e^{-x}
$$

Using boundary conditions:

1. \( y(0) = 2 \Rightarrow C_1 + C_2 = 2 \)
2. \( y'(0) = -1 \Rightarrow -C_2 = -1 \Rightarrow C_2 = 1 \)

Thus, \( C_1 = 1 \), so:

\[
\boxed{y(x) = 1 + e^{-x}}
\]

---

### **(b)**  
$$ y''(x) - y(x) = 0, \quad \text{with boundary conditions } y(0) = 2, \, y'(0) = 0 $$

### **Solution:**
Solve the characteristic equation:

$$
r^2 - 1 = 0
$$

$$
r = \pm 1
$$

General solution:

$$
y(x) = C_1 e^x + C_2 e^{-x}
$$

Using boundary conditions:

1. \( y(0) = 2 \Rightarrow C_1 + C_2 = 2 \)
2. \( y'(0) = 0 \Rightarrow C_1 e^0 - C_2 e^0 = 0 \Rightarrow C_1 = C_2 \)

Thus, \( 2C_1 = 2 \Rightarrow C_1 = 1, C_2 = 1 \), so:

\[
\boxed{y(x) = e^x + e^{-x}}
\]

---

### **(c)**  
$$ \frac{d^2 y(x)}{dx^2} = -\omega^2 y(x) $$

### **Solution:**
Solve the characteristic equation:

$$
r^2 + \omega^2 = 0
$$

$$
r = \pm i\omega
$$

Thus, the general solution:

$$
y(x) = C_1 \cos(\omega x) + C_2 \sin(\omega x)
$$

\[
\boxed{y(x) = C_1 \cos(\omega x) + C_2 \sin(\omega x)}
\]

## **4. Check if the function**  
$$ \psi(t,x) = A\cos(\omega t + kx) $$  
**is a solution of the wave equation:**

$$
\frac{\partial^2 \psi(t,x)}{\partial t^2} - v^2 \frac{\partial^2 \psi(t,x)}{\partial x^2} = 0.
$$

where \( v = \frac{\omega}{k} = \frac{2\pi / T}{2\pi / \lambda} \).

### **Solution:**
Compute the second partial derivatives:

1. **Time derivative**:

$$
\frac{\partial^2 \psi}{\partial t^2} = -A\omega^2 \cos(\omega t + kx)
$$

2. **Spatial derivative**:

$$
\frac{\partial^2 \psi}{\partial x^2} = -A k^2 \cos(\omega t + kx)
$$

Substituting in the wave equation:

$$
-A \omega^2 \cos(\omega t + kx) - v^2 (-A k^2 \cos(\omega t + kx)) = 0
$$

Since \( v = \frac{\omega}{k} \), then \( v^2 k^2 = \omega^2 \), so:

$$
- A \omega^2 \cos(\omega t + kx) + A \omega^2 \cos(\omega t + kx) = 0
$$

Thus, the function **satisfies the wave equation**.

\[
\boxed{\text{Verified}}
\]
