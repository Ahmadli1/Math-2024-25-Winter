## **6. The profit function is**  
$$ P(u) = -2u^2 + 50u - 300 $$  
where \( u \) is the number of units sold. Find the number of units that maximize profit.

### **Solution:**
To find the maximum, take the **first derivative**:

$$
P'(u) = \frac{d}{du} (-2u^2 + 50u - 300)
$$

$$
P'(u) = -4u + 50
$$

Setting \( P'(u) = 0 \) to find critical points:

$$
-4u + 50 = 0
$$

$$
4u = 50 \quad \Rightarrow \quad u = \frac{50}{4} = 12.5
$$

Since the second derivative:

$$
P''(u) = -4
$$

is **negative**, the function has a **maximum** at \( u = 12.5 \).

\[
\boxed{u = 12.5}
\]

## **7. You have 10 meters of string, and you need to use it to enclose the largest possible rectangle. Find the dimensions of the rectangle.**

### **Solution:**
Let **\( x \)** be the width and **\( y \)** be the length of the rectangle. The **perimeter equation** is:

$$
2x + 2y = 10
$$

Solving for \( y \):

$$
y = \frac{10 - 2x}{2} = 5 - x
$$

The **area** function:

$$
A(x) = x y = x(5 - x) = 5x - x^2
$$

Find the **maximum** by differentiating:

$$
A'(x) = 5 - 2x
$$

Setting \( A'(x) = 0 \):

$$
5 - 2x = 0
$$

$$
x = \frac{5}{2} = 2.5
$$

Substituting into \( y \):

$$
y = 5 - 2.5 = 2.5
$$

Thus, the **optimal dimensions** are:

\[
\boxed{x = 2.5, \quad y = 2.5}
\]

## **8. Find extremum of**  
$$ f(x) = x^2 + 3x - 5 $$

### **Solution:**
Differentiate:

$$
f'(x) = 2x + 3
$$

Setting \( f'(x) = 0 \):

$$
2x + 3 = 0
$$

$$
x = -\frac{3}{2}
$$

The **second derivative**:

$$
f''(x) = 2 > 0
$$

Since \( f''(x) > 0 \), the function has a **minimum** at \( x = -\frac{3}{2} \).

Compute \( f(-3/2) \):

$$
f\left(-\frac{3}{2}\right) = \left(-\frac{3}{2}\right)^2 + 3\left(-\frac{3}{2}\right) - 5
$$

$$
= \frac{9}{4} - \frac{9}{2} - 5 = \frac{9}{4} - \frac{18}{4} - \frac{20}{4} = \frac{-29}{4}
$$

Thus, the **minimum point** is:

\[
\boxed{\left( -\frac{3}{2}, -\frac{29}{4} \right)}
\]

## **9. Find extremum of**  
$$ f(x) = \frac{x^2 + 2x + 1}{x - 1} $$

### **Solution:**
Rewrite the function by factoring:

$$
f(x) = \frac{(x+1)^2}{x-1}
$$

Differentiate using the **quotient rule**:

$$
f'(x) = \frac{(2(x+1))(x-1) - (x+1)^2(1)}{(x-1)^2}
$$

Factor the numerator:

$$
f'(x) = \frac{2(x+1)(x-1) - (x+1)(x+1)}{(x-1)^2}
$$

Factor out \( (x+1) \):

$$
f'(x) = \frac{(x+1)[2(x-1) - (x+1)]}{(x-1)^2}
$$

Simplify:

$$
= \frac{(x+1)[2x - 2 - x - 1]}{(x-1)^2}
$$

$$
= \frac{(x+1)(x - 3)}{(x-1)^2}
$$

Setting \( f'(x) = 0 \):

$$
(x+1)(x-3) = 0
$$

Solving for \( x \):

$$
x = -1 \quad \text{or} \quad x = 3
$$

Check the second derivative test:

$$
f''(x) = \frac{(x-1)^2(2x - 2) - (x+1)(x-3)2(x-1)}{(x-1)^4}
$$

Evaluating at \( x = -1 \) and \( x = 3 \):

- **\( x = -1 \)**: The function has a **local minimum**.
- **\( x = 3 \)**: The function has a **local maximum**.

Compute function values:

$$
f(-1) = \frac{(-1)^2 + 2(-1) + 1}{-1 - 1} = \frac{1 - 2 + 1}{-2} = \frac{0}{-2} = 0
$$

$$
f(3) = \frac{3^2 + 2(3) + 1}{3 - 1} = \frac{9 + 6 + 1}{2} = \frac{16}{2} = 8
$$

Thus, the extremum points are:

\[
\boxed{\text{Minimum at } (-1, 0), \quad \text{Maximum at } (3, 8)}
\]
