$$
\int (x^2 + 2) \,dx
$$
### **Solution:**
Using the power rule:

$$
\int x^2 \,dx + \int 2 \,dx = \frac{x^3}{3} + 2x + C
$$

\[
\boxed{\frac{x^3}{3} + 2x + C}
\]

\int 2\sin(x) \,dx
$$

$$
2 \int \sin(x) \,dx = -2\cos(x) + C
$$

\[
\boxed{-2\cos(x) + C}
\]

### **(d)**
$$
\int_1^3 x^2 \,dx
$$
### **Solution:**
Evaluating:

$$
\int x^2 \,dx = \frac{x^3}{3}
$$

$$
\left[ \frac{x^3}{3} \right]_1^3 = \frac{3^3}{3} - \frac{1^3}{3} = \frac{27}{3} - \frac{1}{3} = \frac{26}{3}
$$

\[
\boxed{\frac{26}{3}}
\]

### **(h)**
$$
\int \sqrt{x} \,dx
$$
### **Solution:**
Rewriting \( \sqrt{x} = x^{1/2} \):

$$
\int x^{1/2} \,dx = \frac{x^{3/2}}{3/2} = \frac{2}{3} x^{3/2} + C
$$

\[
\boxed{\frac{2}{3} x^{3/2} + C}
\]

---

### **(i)**
$$
\int \sqrt{10x} \,dx
$$
### **Solution:**
Substituting \( u = 10x \),

$$
\int \sqrt{10x} \,dx = \frac{2}{3} (10x)^{3/2} \times \frac{1}{10}
$$

$$
= \frac{2}{3} \times \frac{1}{10} (10x)^{3/2} + C = \frac{2}{30} (10x)^{3/2} + C
$$

\[
\boxed{\frac{2}{30} (10x)^{3/2} + C}
\]

---

### **(j)**
$$
\int \cos \left(\frac{5}{2}x + 3\right) \,dx
$$
### **Solution:**
Using substitution \( u = \frac{5}{2}x + 3 \), \( du = \frac{5}{2}dx \):

$$
\int \cos u \cdot \frac{2}{5} \,du = \frac{2}{5} \sin u + C
$$

\[
\boxed{\frac{2}{5} \sin \left(\frac{5}{2}x + 3\right) + C}
\]

---

### **(k)**
$$
\int \frac{\cos(\ln x)}{x} \,dx
$$
### **Solution:**
Using substitution \( u = \ln x \), \( du = \frac{dx}{x} \):

$$
\int \cos u \,du = \sin u + C
$$

\[
\boxed{\sin(\ln x) + C}
\]

---

### **(l)**
$$
\int x \ln x \,dx
$$
### **Solution:**
Using integration by parts, \( u = \ln x \), \( dv = xdx \):

$$
\int x \ln x \,dx = \frac{x^2}{2} \ln x - \frac{x^2}{4} + C
$$

\[
\boxed{\frac{x^2}{2} \ln x - \frac{x^2}{4} + C}
\]

---

### **(m)**
$$
\int x e^x \,dx
$$
### **Solution:**
Using integration by parts:

$$
\int x e^x \,dx = x e^x - \int e^x \,dx = x e^x - e^x + C
$$

\[
\boxed{x e^x - e^x + C}
\]

## **3. Calculate the area of the region bounded by the lines**  
\( x = 1 \), \( x = 2 \), \( y = 0 \), and \( y = x^2 + 1 \).

### **Solution:**
The area is:

$$
A = \int_1^2 (x^2 + 1) \,dx
$$

Computing:

$$
\left[ \frac{x^3}{3} + x \right]_1^2 = \left(\frac{8}{3} + 2\right) - \left(\frac{1}{3} + 1\right) = \frac{8}{3} + 2 - \frac{1}{3} - 1
$$

$$
= \frac{7}{3} + 1 = \frac{10}{3}
$$

\[
\boxed{\frac{10}{3}}
\]
## **4. Calculate the area under the sine curve over the interval** \( [0,\pi] \).

### **Solution:**
The area under the sine curve is given by:

$$
P = \int_0^\pi \sin(x) \,dx
$$

Computing the integral:

$$
\int \sin(x) \,dx = -\cos(x)
$$

Evaluating from \( 0 \) to \( \pi \):

$$
\left[ -\cos(x) \right]_0^\pi = -\cos(\pi) + \cos(0) = -(-1) + 1 = 1 + 1 = 2
$$

\[
\boxed{2}
\]

## **5. Calculate the length of the sine curve over the same interval** \( [0,\pi] \).

### **Solution:**
The arc length formula is:

$$
L = \int_a^b \sqrt{1 + (f'(x))^2} \,dx
$$

For \( f(x) = \sin(x) \), we compute its derivative:

$$
f'(x) = \cos(x)
$$

Thus, the arc length integral is:

$$
L = \int_0^\pi \sqrt{1 + \cos^2(x)} \,dx
$$

This integral does not have a simple closed-form solution, but it can be **computed numerically**.

\[
\boxed{L = \int_0^\pi \sqrt{1 + \cos^2(x)} \,dx}
\]

## **6. Find the distance of the moving particle between time \( t = 0 \) and \( t = 2 \) for the following position function:**  
$$ x(t) = 3t^2 - 6t + 1 $$

### **Solution:**
The velocity function is:

$$
V(t) = x'(t) = \frac{d}{dt} (3t^2 - 6t + 1) = 6t - 6
$$

The total distance traveled is:

$$
D = \int_0^2 |V(t)| \,dt
$$

First, find when \( V(t) = 0 \):

$$
6t - 6 = 0 \quad \Rightarrow \quad t = 1
$$

Since \( V(t) \) changes sign at \( t = 1 \), we split the integral:

$$
D = \int_0^1 |6t - 6| \,dt + \int_1^2 |6t - 6| \,dt
$$

For \( 0 \leq t \leq 1 \), \( 6t - 6 \leq 0 \), so \( |6t - 6| = -(6t - 6) = -6t + 6 \).

For \( 1 \leq t \leq 2 \), \( 6t - 6 \geq 0 \), so \( |6t - 6| = 6t - 6 \).

Computing both integrals:

$$
\int_0^1 (-6t + 6) \,dt = \left[ -3t^2 + 6t \right]_0^1 = (-3(1)^2 + 6(1)) - (-3(0)^2 + 6(0))
$$

$$
= (-3 + 6) - 0 = 3
$$

$$
\int_1^2 (6t - 6) \,dt = \left[ 3t^2 - 6t \right]_1^2
$$

$$
= (3(2)^2 - 6(2)) - (3(1)^2 - 6(1))
$$

$$
= (12 - 12) - (3 - 6) = 0 + 3 = 3
$$

Total distance:

$$
D = 3 + 3 = 6
$$

\[
\boxed{6}
