## **1. Find the Taylor series and visualize obtained functions in Geogebra:**

### **(a) \( f(x) = \cos(x) \) around \( x = 0 \) up to the 4th degree.**

The Taylor series expansion of \( \cos(x) \) at \( x = 0 \) is:

$$
\cos(x) = \sum_{n=0}^{\infty} \frac{(-1)^n}{(2n)!} x^{2n}
$$

Expanding up to the 4th degree:

$$
\cos(x) \approx 1 - \frac{x^2}{2!} + \frac{x^4}{4!}
$$

$$
\cos(x) \approx 1 - \frac{x^2}{2} + \frac{x^4}{24}
$$

\[
\boxed{\cos(x) \approx 1 - \frac{x^2}{2} + \frac{x^4}{24}}
\]

### **(b) \( h(x) = \frac{1}{1 - x} \) around \( x = 0 \) up to the 4th degree.**

The Taylor series expansion of \( \frac{1}{1 - x} \) at \( x = 0 \) is:

$$
h(x) = \sum_{n=0}^{\infty} x^n
$$

Expanding up to the 4th degree:

$$
h(x) \approx 1 + x + x^2 + x^3 + x^4
$$

\[
\boxed{h(x) \approx 1 + x + x^2 + x^3 + x^4}
\]

### **(c) \( g(x) = \sin(x) \) around \( x = \pi \) up to the 4th degree.**

The Taylor series expansion of \( \sin(x) \) at \( x = a \) is:

$$
\sin(x) = \sum_{n=0}^{\infty} \frac{(-1)^n}{(2n+1)!} (x-a)^{2n+1}
$$

Expanding around \( x = \pi \), up to the 4th degree:

$$
\sin(x) \approx \sin(\pi) + \cos(\pi)(x - \pi) - \frac{\sin(\pi)}{2!} (x - \pi)^2 - \frac{\cos(\pi)}{3!} (x - \pi)^3
$$

Since \( \sin(\pi) = 0 \) and \( \cos(\pi) = -1 \), we get:

$$
\sin(x) \approx 0 - (x - \pi) - 0 + \frac{(x - \pi)^3}{6}
$$

\[
\boxed{\sin(x) \approx - (x - \pi) + \frac{(x - \pi)^3}{6}}
\]

## **2. Find a tangent line to the function \( f(x) = e^{\sin(x)} \) at \( x_0 = \pi \).**

### **Solution:**
The **tangent line equation** is:

$$
y = f'(x_0)(x - x_0) + f(x_0)
$$

### **Step 1: Compute \( f(x) = e^{\sin(x)} \) at \( x_0 = \pi \):**

$$
f(\pi) = e^{\sin(\pi)} = e^0 = 1
$$

### **Step 2: Compute \( f'(x) \):**

Using the chain rule:

$$
f'(x) = e^{\sin(x)} \cos(x)
$$

Evaluate at \( x = \pi \):

$$
f'(\pi) = e^{\sin(\pi)} \cos(\pi) = e^0 (-1) = -1
$$

### **Step 3: Equation of the tangent line:**

$$
y = -1(x - \pi) + 1
$$

\[
\boxed{y = -x + \pi + 1}
\]
